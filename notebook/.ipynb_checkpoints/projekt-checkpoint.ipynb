## Ekstremne vrijednosti

```python
# traženje ekstremnih vrijednosti za varijablu HeatingLoad  

GG_HL = df.HeatingLoad.mean() + (df.HeatingLoad.std()*3)
DG_HL = df.HeatingLoad.mean() - (df.HeatingLoad.std()*3)

if df.HeatingLoad.max() > GG_HL and df.HeatingLoad.min() < DG_HL:
    print('Postoje ekstremne vrijednosti u varijabli HeatingLoad!')
else:
    print('Ne postoje ekstremne vrijednosti u varijabli HeatingLoad.')

# traženje ekstremnih vrijednosti za varijablu CoolingLoad
  
GG_CL = df.CoolingLoad.mean() + (df.CoolingLoad.std()*3)
DG_CL = df.CoolingLoad.mean() - (df.CoolingLoad.std()*3)

if df.CoolingLoad.max() > GG_CL and df.CoolingLoad.min() < DG_CL:
    print('Postoje ekstremne vrijednosti u varijabli CoolingLoad!')
else:
    print('Ne postoje ekstremne vrijednosti u varijabli CoolingLoad.')
```

# Univarijantna analiza

## Ciljna varijabla

Prije daljnjih radnji treba odrediti ciljnu varijablu koja određuje koji problem će model rješavati tj. pomoću koje će se model trenirati i testirat.

Budući da se radi o podacima vezanim uz dimenzije zgrade, ostakljenost, potrošnji energije i sl. ciljna varijabla će govoriti je li pojedina zgrada energetski učinkovita ili ne.

Postoje razredi energetske učikovitosti npr. za stanove postoje A+, A, B, itd., a za ciljnu varijablu je potrebno samo označiti da li zgrada je ili nije energetski učinkovita.

B razred bi se mogao uzeti kao osnova za razlikovanje energetski učinkovitih jer se većinom B razred gleda. Ako je u pitanju B razred po specifikaciji koju sam našao na Internetu potrošnja mora biti manje od 50 W, a maksimalna vrijednost u datasetu za grijanje 43.1 te za hlađenje 48.03 što onda nema smisla jer bi ispalo da su sve zgrade energetski učinkovite.

Kada su u pitanju poslovne zgrade i postrojenja, kolega koji se time prije bavio mi je rekao da je to varijabilno sve i da ne postoji strogo definirano n W / m^3 kao minimalna granica da bi nešto prošlo kao energetski učinkovito.
Poželjno svake godine smanjiti potrošnju kroz razne načine npr. boljim izolacijskim materijalima (o čemu nema podataka u ovome datasetu).

Kao što je bilo vidljivo iz prijašnjih grafova, varijable HeatingLoad i CoolingLoad imaju bimodalnu distribuciju koja dijeli skupinu na energetski učinkovitu i ona koja nije energetski učinkovita, ali ovo je jako uvjetovano jednom varijablom, a to je OverallHeight. Iz tog razloga nebi bilo dobro uzeti preciznu granicu između njih koja je jako uvjetovana jednom varijablom koja je u velikoj korelaciji s HeatingLoad i CoolingLoad tj. sa varijablom potrosnja.

Iz navedenih razloga kriterij po kojem će ciljna varijabla biti postavljena je sljedeći:

Polje potrošnja biti će sortirano prema veličini (od najmanje do najveće vrijednosti) i manja polovica elemenata biti će označena kao energetski učinkovita.

Kriterij je mogao biti i stroži npr. da je granica samo 30% ili 40% najnižih vrijednosti klasificirano kao energetski učinkovito ali ja osobno smatram da je to prestrogo i da nema smisla da je energetska učinkovitost definirana vrijednosti koja je manja od cca 16 W, gdje je maksimalna vrijednost varijable potrosnja 44.98.

Od donjih 50% vrijednosti potrošnje, maksimalna vrijednost je 20.48 što je i dalje manje od 50% globalne maksimalne vrijednosti.

### Druga iteracija definiranja ciljne varijable:

IV je pokazao da je najpovoljniji kriterij od 35% donjih vrijednosti potrosnje.

Inače, ako je kriterij 50% ili sve niže od 22 W po potrosnji što dijeli bimodalnu distribuciju, prediktivnost je s WallArea i GlazingArea samo oko 55%.


U dosadašnjim modelima nije bilo dummy varijabli iz razloga što je IV bio u redu samo za 1 varijablu (GlazingArea), ako ćemo biti malo manje strogi i dopustiti IV od 1,04 za varijablu WallArea onda su bile 2 varijable koje su u redu po IV.

Ovime sam htio vidjeti kako će se logistička regresija ponašati bez dummy varijabli.

```python
potrosnja = list(df['potrosnja'])
potrosnja.sort(reverse=False)
potrosnja = potrosnja[:round(len(potrosnja)*0.35)]
#print(potrosnja)
kriterij = potrosnja[-1]
print('Potrošnja <= {0} smatrat će se učinkovitom.'.format(kriterij))

df['ciljna'] = False
df.loc[df.potrosnja <= kriterij, 'ciljna'] = True 
df.head(10)
učinkoviti = df.ciljna.loc[df.ciljna == True].count()
print('Postoji {0} zgrada prema ovom kriteriju.'.format(učinkoviti))
```

## Story telling - opis problema

```python
for x in df.OverallHeight.unique():
    print('za OverallHeight {0}: '.format(x))
    print(df['potrosnja'].loc[df['OverallHeight'] == x].mean())
    print('')
```

Varijabla OverallHeight ima veliki utjecaj na potrošnju i time je kandidat za model. 

Što je zgrada viša to je potrošnja energije veća iz razloga što je shodno time i veća površina zidova izloženija vanjskoj temeraturi, a površina podloge koja leži na zemlji je manja i time nema toliko veliki benefit.

Kada je zgrada niža onda je dulja i šira te shodno time je i površina podloge veća, ovime su površine zidova manje i utjecaj valjnjskih temeratura je na manju površinu za razliku od prethnodnog slučaja.

Budući da je riječ o varijabli koja ima samo dvije vrijednosti i postoji ogromna razlika između srednjih vrijednosti moguće je da će se dogodit overfitting modela zbog toga što se bi se model svodio na logiku if grananja tj. ako je zgrada jednokatnica (3.5) onda je energetski učinkovita zgrada, a ako je dvokatnica (7.0) onda nije učinkovita, što nije istina. Visina naravno utječe na energetsku učinkovitost, ali postoje primjeri kod kojih je riječ o dvokatnicama, a da je njihova potrošnja manja od najveće potrošnje jednokatnice koja je označena kao energetski učinkovita.

```python
print('Energetski učinkovite dvokatnice:')
print(df[['potrosnja', 'OverallHeight', 'ciljna']].loc[(df['potrosnja'] < kriterij) & (df['OverallHeight'] == 7.0)])
print('')
print('Najveća potrošnja energetski učinkovite jednokatnice:')
print(df[['potrosnja', 'OverallHeight', 'ciljna']].loc[(df['potrosnja'] < kriterij) & (df['OverallHeight'] == 3.5)].max())
```

Budući da je u drugoj iteraciji kriterij snižen s 50% na 35% više ne postoje zgrade s OverallHeight-om od 7.0 koje su označene kao energetski učinkovite.

## WoE IV

### RelativeCompactness

```python
RelativeCompactness_bin = pd.qcut(df.RelativeCompactness,3)
#print(RelativeCompactness_bin)
df['RC_bin']=RelativeCompactness_bin
df['RC_bin'] = df['RC_bin'].astype(str)
df.info()
```

```python
df['RC_bin'].describe()
```

```python
RC = pd.crosstab(df['RC_bin'], df['ciljna']).apply(lambda c: c/c.sum(), axis=0)

RC[False] = RC[False].replace(0, 0.000001)
RC[True] = RC[True].replace(0, 0.000001)
```

```python
RC
```

```python
RC['WoE'] = np.log(RC[True]/RC[False])
RC['IV'] = (RC[True]-RC[False])* np.log(RC[True]/RC[False])
print(RC['IV'].sum())
RC
```

Information value je ogroman zbog visoke koreliranosti

```python
scipy.stats.pearsonr(df['RelativeCompactness'], df['potrosnja'])
```

```python
RC['WoE'].plot(kind='bar')
plt.show()
```

Varijabla ima prevelik IV i neće se koristiti kao prediktor.

## Selekcija najjačih varijabli

Samim time je redudantna informacija jer imamo te 2 varijable. I neće se koristiti.

Relative compactness je isto redudantna varijabla jer je ona obrnuto proporcionalna sa SurfaceArea i ima istu (visoku) korelaciju sa varijablom potrosnja i ciljna samo je negativna korelacija. Neće se koristiti.

OverallHeight će overfittat model, postoje samo 2 moguće vrijednosti i varijabla je u visokoj korelaciji s potrosnjom i s ciljnom.

Orientation ima mali IV.

GlazingAreaDistribution ima mali IV.

Varijable koje će se koristiti u prvom modelu su:

WallArea i dummy varijable nastale iz GlazingArea koje su grupirane prema WoE (imamo 3 dummy varijable).

# Razvoj modela

## Logistička regresija i proračun prediktivnosti

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import roc_curve, auc
import random
```

```python
train = df[[  'ostakljenost_visoka', 'ostakljenost_umjerena', 'neostakljeno', 'WallArea', 'ciljna']]
X = train.iloc[:,:-1]
y = train.iloc[:,-1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)

regression = LogisticRegression(fit_intercept=True)
regression.fit(X_train, y_train)

training_pred = regression.predict(X_train)

print('Prediktivnost na training uzorku: {:.2f}%'.format(regression.score(X_train, y_train) * 100))

testing_pred = regression.predict(X_test)

print('Prediktivnost na testing uzorku: {:.2f}%'.format(regression.score(X_test, y_test) * 100))

actual_test = y_test
predictions_test = testing_pred
false_positive_rate, true_positive_rate, thresholds = roc_curve(actual_test, predictions_test)
roc_auc = auc(false_positive_rate, true_positive_rate)

plt.title('Receiver Operating Characteristic (ROC) curve')
plt.plot(false_positive_rate, true_positive_rate, 'b',
label='AUC = %0.2f'% roc_auc)
plt.legend(loc='lower right')
plt.plot([0,1],[0,1],'r--')
plt.xlim([-0.1,1.1])
plt.ylim([-0.1,1.1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

Drugi model je napravljen s varijablom GlazingArea kao kontinuiranom, umjesto izvedenu kao 3 dummy varijable, što je pokazalo veću prediktivnost.

```python
train = df[[  'GlazingArea', 'WallArea', 'ciljna']]
X = train.iloc[:,:-1]
y = train.iloc[:,-1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)

regression = LogisticRegression(fit_intercept=True)
regression.fit(X_train, y_train)

training_pred = regression.predict(X_train)

print('Prediktivnost na training uzorku: {:.2f}%'.format(regression.score(X_train, y_train) * 100))

testing_pred = regression.predict(X_test)

print('Prediktivnost na testing uzorku: {:.2f}%'.format(regression.score(X_test, y_test) * 100))

actual_test = y_test
predictions_test = testing_pred
false_positive_rate, true_positive_rate, thresholds = roc_curve(actual_test, predictions_test)
roc_auc = auc(false_positive_rate, true_positive_rate)

plt.title('Receiver Operating Characteristic (ROC) curve')
plt.plot(false_positive_rate, true_positive_rate, 'b',
label='AUC = %0.2f'% roc_auc)
plt.legend(loc='lower right')
plt.plot([0,1],[0,1],'r--')
plt.xlim([-0.1,1.1])
plt.ylim([-0.1,1.1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

Očeivao sam od ove varijable budući da je imala i korelaciju cca 0,7 da će doći do overfittinga ali preciznost na training setu je manja od one na testing setu tako da nema overfittinga.

Razlog zašto sam odbrao RoofArea a ne neku drugu varijablu je taj što je RoofArea bio u korelaciji s OverallHeight (koji je potencijalno dobar prediktor međutim mogao bi izazvati overfitting), OverallHeight ima samo 2 modaliteta za razliku od RoofArea koji ima više i zato sam odlučio pokušati i s RoofArea.

```python
train = df[[ 'GlazingArea', 'WallArea', 'RoofArea', 'ciljna']]
X = train.iloc[:,:-1]
y = train.iloc[:,-1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)

regression = LogisticRegression(fit_intercept=True)
regression.fit(X_train, y_train)

training_pred = regression.predict(X_train)

print('Prediktivnost na training uzorku: {:.2f}%'.format(regression.score(X_train, y_train) * 100))

testing_pred = regression.predict(X_test)

print('Prediktivnost na testing uzorku: {:.2f}%'.format(regression.score(X_test, y_test) * 100))

actual_test = y_test
predictions_test = testing_pred
false_positive_rate, true_positive_rate, thresholds = roc_curve(actual_test, predictions_test)
roc_auc = auc(false_positive_rate, true_positive_rate)

plt.title('Receiver Operating Characteristic (ROC) curve')
plt.plot(false_positive_rate, true_positive_rate, 'b',
label='AUC = %0.2f'% roc_auc)
plt.legend(loc='lower right')
plt.plot([0,1],[0,1],'r--')
plt.xlim([-0.1,1.1])
plt.ylim([-0.1,1.1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

Ovaj model, zadnji, ima iste performanse (ista prediktivnost i na training i na testing setu i također ista RoC krivulja i AUC) ako se varijabla RoofArea zamjeni sa SurfaceArea budući da su jako korelirane (0.88), a i SurfaceArea nastaje formulom koja je objašnjena ranije.

Drugim riječima imamo redudantnu varijablu koja se samo drugačije zove.

## Izračun HHI indeksa na uzorku za razvoj i testiranje

```python
# spajanje prediktora i ciljne za test
df_test = pd.DataFrame(X_train)
df_test['ciljna'] = y_train
df_test['train_test'] = 'test'

# spajanje prediktora i ciljne za training
df_train = pd.DataFrame(X_test)
df_train['ciljna'] = y_test
df_train['train_test'] = 'train'

# spajanje training i testing setova
frames = [df_train, df_test]
df_final = pd.concat(frames)
```

```python
HH_ciljna = pd.crosstab(df_final.ciljna, df_final.train_test).apply(lambda c: pow(c/c.sum(),2), axis=0)
print(HH_ciljna)
HH_ciljna.sum()
```

Koncentracija ukazuje na to da je manje True vrijednosti u ciljnoj tj. da je manje zgrada koje su energetski učinkovite nego zgrada koje nisu energetski učinkovite što je i za očekivati zbog kriterija koji je uzimao 35% nižih vrijednosti po potrošnji.

```python
HH_GA = pd.crosstab(df_final.GlazingArea, df_final.train_test).apply(lambda c: pow(c/c.sum(),2), axis=0)
print(HH_GA)
HH_GA.sum()
```

Zgrade koje nisu ostakljene najmanje su koncentrirane.

```python
HH_WA = pd.crosstab(df_final.WallArea, df_final.train_test).apply(lambda c: pow(c/c.sum(),2), axis=0)
print(HH_WA)
HH_WA.sum()
```

```python
HH_RA = pd.crosstab(df_final.RoofArea, df_final.train_test).apply(lambda c: pow(c/c.sum(),2), axis=0)
print(HH_RA)
HH_RA.sum()
```

Kod RoofArea koncentracija ukazuje da je više zgrada koje imaju krov 220,50.

Koncentracije kod testing i training setova su podjednake što i je poželjno (da su prevelike razlike trebalo bi ponovo raditi uzorkovanje), ne smiju se značajno razlikovati jer bi moglo utjecati na model.

## Izračun indeksa stabilnosti između uzorka za razvoj i testiranje na vjerojatnostnoj distribuciji

```python
PSI_ciljna = pd.crosstab(df_final.ciljna,df_final.train_test).apply(lambda c: c/c.sum(), axis=0)
PSI_ciljna

PSI_ciljna['PSI'] = (PSI_ciljna['train']-PSI_ciljna['test'])* np.log(PSI_ciljna['train']/PSI_ciljna['test'])
print('Indeks stabilnosti za ciljnu:')
print(PSI_ciljna['PSI'].sum())
```

```python
PSI_WA = pd.crosstab(df_final.WallArea,df_final.train_test).apply(lambda c: c/c.sum(), axis=0)
PSI_WA

PSI_WA['PSI'] = (PSI_WA['train']-PSI_WA['test'])* np.log(PSI_WA['train']/PSI_WA['test'])
print('Indeks stabilnosti za WallArea:')
print(PSI_WA['PSI'].sum())
```

```python
PSI_RA = pd.crosstab(df_final.RoofArea,df_final.train_test).apply(lambda c: c/c.sum(), axis=0)
PSI_RA

PSI_RA['PSI'] = (PSI_RA['train']-PSI_RA['test'])* np.log(PSI_RA['train']/PSI_RA['test'])
print('Indeks stabilnosti za RoofArea:')
print(PSI_RA['PSI'].sum())
```

```python
PSI_GA = pd.crosstab(df_final.GlazingArea,df_final.train_test).apply(lambda c: c/c.sum(), axis=0)
PSI_GA

PSI_GA['PSI'] = (PSI_GA['train']-PSI_GA['test'])* np.log(PSI_GA['train']/PSI_GA['test'])
print('Indeks stabilnosti za GlazingArea:')
print(PSI_GA['PSI'].sum())
```

Kod svih varijabli je indeks stabilnosti ispod 0,1 što znači da su uzorci stabilni.
